## GMM implementation

Step 1: Parameter initialisation

- For the mean mu, since the initialisation direct influences the probability distribution, we want our mean to be within reasonable range such that for each component, each point of that component stays within the pdf of the distribution
  

idea 1: Initialise randomly.........flopped

idea 2: Initialise with fuzzy c means........flopped

idea 3: initialise with random points of our dataset........ closs enough but very erratic........flopped

idea 4: initialise with k-Means......works well. But needs fine tuning of algorithm


- For sigma the covariance matrix, the best option was to get the general idea of variance within our dataset. So for each component the covariance matrix of the dataset was used

```python
# perform GMM
#initialise parameters
def init(data, n_clusters):
    from sklearn.cluster import KMeans
    from scipy.stats import random_correlation
    seed = 42
    high = data.max()
    low = data.min()
    rng = np.random.default_rng(seed)
    s_eig = rng.random(data.shape[1])
    s_eig = s_eig / np.sum(s_eig) * data.shape[1]
    cov = random_correlation.rvs(s_eig, random_state=rng)
    km = KMeans(n_clusters).fit_predict(data)
    index = []
    for i in np.unique(km):
        ind=  np.flatnonzero(km == i)
        index.append(ind)
    c = []
    for j in range(len(np.unique(km))):
        c.append(np.random.choice(index[j], 1, replace=False))
    N = np.array(c).astype(int).flatten()
    mu = data.to_numpy()[N]
    #mu = km.cluster_centers_
    #_,mu,_ = fcm(n_clusters,2,data.to_numpy(), 100)
    
    #mu = [rng.uniform(low = low, high = high, size = [10]) for _ in range(n_clusters)]
    #random_indices = np.random.randint(0, data.shape[0], n_clusters)
    #mu = data.to_numpy()[random_indices,:]#center of distribution
    sigma = [np.cov(data.T)+1 for _ in range(n_clusters)]#covariance matrix
    pi = np.full(shape=n_clusters, fill_value=1/n_clusters)
    #pi, weights of each component
    #responsibility = np.full(shape = data.shape, fill_value = 1/n_clusters)
    return mu, sigma,pi#, random_indices
```

Step 2: Expectation Step alias calculating the responsibility matrix

```python
def e_step(data, n_clusters, mu, sigma, pi):
    likelihood = np.zeros((data.shape[0], n_clusters))
    #responsibility = np.zeros_like(likelihood)
    for i in range(n_clusters):
        mui = mu[i]
        covi = sigma[i] 
        if any(np.isnan(mui)):
            new_mu, new_sigma, new_pi = init(data, n_clusters)
            mui = new_mu[i]
            covi = new_sigma[i]
        

        #normal_dist = scipy.stats.multivariate_normal(mean = mu[i], cov = sigma[i])
        likelihood[:,i] = scipy.stats.multivariate_normal( mean = mui, cov = covi, allow_singular=True).pdf(data)
    numerator = (likelihood*pi) + 1e-18
    denominator = numerator.sum(axis = 1)[:, np.newaxis] 
    denominator = denominator 
    

    responsibility = numerator/denominator
    return responsibility
```

Step 3: Updating parameters with the Maximization Step

```python
def m_step(data,mu, sigma, pi, n_clusters, responsibility):
    cov = np.zeros_like(sigma)

    for i in range(n_clusters):
        weight = responsibility[:, [i]]
        total_weight = weight.sum()
        mu[i] = (data * weight).sum(axis=0) / total_weight
        
        #cov[i] = ((data - mu[i]).T @ (data- mu[i]) * (responsibility[:,i]).sum())/(responsibility[:,i].sum())
        cov[i] = np.cov(data.T,aweights=(weight/total_weight).flatten(), bias=True)
        pi[i] = responsibility[:,i].sum()/len(data)
    return np.array(mu), np.array(pi), np.array(cov)
```

Step 4: Evaluation of the loglikelihood 

```python
def log_likelihood(data, n_clusters, pi, sigma, mu, responsibility):
    eps = 1e-8
    #likelihood = np.array([scipy.stats.multivariate_normal(mean = mu, cov = sigma, allow_singular= True).pdf(data) for mu, sigma in zip(mu, sigma)])
    #weighted_likelihood =  np.log(likelihood + eps).T +np.log(pi) 
    #amount = [] 
    #p = weighted_likelihood.sum(0)
    likelihood_loss = []
    for k in range(n_clusters):
        normal_dist = scipy.stats.multivariate_normal(mu[k], sigma[k], allow_singular=True).pdf(data)
        likelihood = pi[k] *normal_dist
        inner = (responsibility[:,k] * np.log(likelihood + eps)).sum()
        likelihood_loss.append(inner.sum())
    return np.sum(-1 * np.array(likelihood_loss))
    #return weighted_likelihood
```

Final Step: EM-Algorithm

```python
def gmm(data, max_iter, n_clusters):
    iter = 0
    mu, sigma, pi= init(data, n_clusters)
    l = []
    while iter < max_iter:
        responsibility = e_step(data, n_clusters, mu, sigma, pi)
        mu, pi, sigma = m_step(data, mu, sigma, pi, n_clusters, responsibility)
        loss = log_likelihood(data, n_clusters,pi,sigma,mu,responsibility)
        l.append(loss)
        
        iter += 1
        #print(f'loss: {loss}')
        plt.plot(l)
        plt.xlabel('Number of Iterations')
        plt.ylabel('log_likelihood')
        plt.title('loss plot')
    return np.argmax(np.array(responsibility), axis = 1)
        
        
```

Implementation of the EM-Algortithm on our dataset

```python
#here we perform muliple gmm takes till our cost function looks like an l shaped smooth curve
labels = gmm(data, 10, 10)
labels

```
